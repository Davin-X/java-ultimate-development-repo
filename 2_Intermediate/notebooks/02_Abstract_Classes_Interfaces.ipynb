## Abstract Classes Fundamentals

**Classes that cannot be instantiated but provide partial implementations for subclasses**

```java
// Abstract classes - building contracts with partial implementations
public class AbstractClassesDemo {
    
    // Abstract class - cannot be instantiated directly
    static abstract class Shape {
        protected String name;
        
        public Shape(String name) {
            this.name = name;
        }
        
        // Abstract method - must be implemented by subclasses
        public abstract double calculateArea();
        
        // Concrete method - provided implementation
        public void display() {
            System.out.println("Shape: " + name);
        }
        
        // Another concrete method
        public void printArea() {
            System.out.println(name + " area: " + calculateArea());
        }
    }
    
    // Concrete implementation of Shape
    static class Circle extends Shape {
        private double radius;
        
        public Circle(double radius) {
            super("Circle");  // Call parent constructor
            this.radius = radius;
        }
        
        // Must implement the abstract method
        @Override
        public double calculateArea() {
            return Math.PI * radius * radius;
        }
    }
    
    static class Rectangle extends Shape {
        private double width, height;
        
        public Rectangle(double width, double height) {
            super("Rectangle");
            this.width = width;
            this.height = height;
        }
        
        @Override
        public double calculateArea() {
            return width * height;
        }
    }
    
    // Abstract class with both abstract and concrete methods
    static abstract class Vehicle {
        protected String brand;
        protected int year;
        
        public Vehicle(String brand, int year) {
            this.brand = brand;
            this.year = year;
        }
        
        // Abstract methods - must be implemented
        public abstract void start();
        public abstract void stop();
        
        // Concrete method - provided
        public void honk() {
            System.out.println(brand + " goes: HONK! HONK!");
        }
        
        public void displayInfo() {
            System.out.println("Vehicle: " + brand + " (" + year + ")");
        }
    }
    
    static class Car extends Vehicle {
        public Car(String brand, int year) {
            super(brand, year);
        }
        
        @Override
        public void start() {
            System.out.println(brand + " car: Turning key... Engine started!");
        }
        
        @Override
        public void stop() {
            System.out.println(brand + " car: Pressing brake... Engine stopped.");
        }
    }
    
    static class Motorcycle extends Vehicle {
        public Motorcycle(String brand, int year) {
            super(brand, year);
        }
        
        @Override
        public void start() {
            System.out.println(brand + " motorcycle: Kickstarting... Vroom!");
        }
        
        @Override
        public void stop() {
            System.out.println(brand + " motorcycle: Killing engine... Stopped.");
        }
    }
    
    public static void demonstrateAbstractClasses() {
        System.out.println("=== ABSTRACT CLASSES DEMONSTRATION ===\n");
        
        // Cannot create Shape objects directly
        // Shape genericShape = new Shape("Generic"); // COMPILATION ERROR!
        
        System.out.println("Creating concrete shape implementations:");
        Circle circle = new Circle(5.0);
        Rectangle rectangle = new Rectangle(4.0, 6.0);
        
        Shape[] shapes = {circle, rectangle}; // Polymorphism!
        
        for (Shape shape : shapes) {
            shape.display();      // Inherited concrete method
            shape.printArea();    // Uses abstract method implementation
            System.out.println();
        }
        
        System.out.println("Vehicle implementations:");
        Car car = new Car("Toyota", 2022);
        Motorcycle bike = new Motorcycle("Honda", 2021);
        
        Vehicle[] vehicles = {car, bike};
        
        for (Vehicle vehicle : vehicles) {
            vehicle.displayInfo(); // Concrete method
            vehicle.start();       // Implemented abstract method
            vehicle.honk();        // Concrete method
            vehicle.stop();        // Implemented abstract method
            System.out.println();
        }
    }
    
    public static void main(String[] args) {
        demonstrateAbstractClasses();
        
        System.out.println("\nðŸŽ¯ ABSTRACT CLASSES SUMMARY:");
        System.out.println("â€¢ Cannot be instantiated directly");
        System.out.println("â€¢ Can have abstract (unimplemented) methods");
        System.out.println("â€¢ Can have concrete (implemented) methods");
        System.out.println("â€¢ Provide partial implementations for subclasses");
        System.out.println("â€¢ Used when classes share common behavior but differ in specifics");
    }
}
```

## Java Interfaces

**Pure contracts that define what classes must implement, without specifying how**

```java
// Java interfaces - pure contracts, multiple inheritance
public class InterfacesDemo {
    
    // Basic interface - defines a contract
    interface Drawable {
        void draw();  // Abstract method - implementation required
        double getArea(); // Another abstract method
    }
    
    // Another interface
    interface Resizable {
        void resize(double factor);
        boolean canResize(double factor);
    }
    
    // Interface with constants
    interface GraphicsConstants {
        double PI = 3.14159;  // Automatically public static final
        int MAX_WIDTH = 1920;  // Automatically public static final
    }
    
    // Class implementing one interface
    static class Circle implements Drawable, GraphicsConstants {
        private double radius;
        
        public Circle(double radius) {
            this.radius = radius;
        }
        
        @Override
        public void draw() {
            System.out.println("Drawing a circle with radius " + radius);
        }
        
        @Override
        public double getArea() {
            return PI * radius * radius;
        }
    }
    
    // Class implementing multiple interfaces (multiple inheritance)
    static class Rectangle implements Drawable, Resizable {
        private double width, height;
        
        public Rectangle(double width, double height) {
            this.width = width;
            this.height = height;
        }
        
        @Override
        public void draw() {
            System.out.println("Drawing a rectangle " + width + "x" + height);
        }
        
        @Override
        public double getArea() {
            return width * height;
        }
        
        @Override
        public void resize(double factor) {
            width *= factor;
            height *= factor;
            System.out.println("Rectangle resized by factor " + factor);
        }
        
        @Override
        public boolean canResize(double factor) {
            return factor > 0; // Only positive scaling allowed
        }
    }
    
    // Interface demonstration
    public static void demonstrateInterfaces() {
        System.out.println("=== JAVA INTERFACES DEMONSTRATION ===\n");
        
        // Create objects
        Circle circle = new Circle(5.0);
        Rectangle rectangle = new Rectangle(4.0, 6.0);
        
        // Use through interface references (polymorphism)
        Drawable[] drawables = {circle, rectangle};
        
        System.out.println("Drawing all shapes:");
        for (Drawable drawable : drawables) {
            drawable.draw();           // Interface method
            System.out.println("Area: " + drawable.getArea());
            System.out.println();
        }
        
        // Test multiple interfaces
        System.out.println("Testing Resizable interface:");
        if (rectangle.canResize(2.0)) {
            rectangle.resize(2.0);
            System.out.println("New area: " + rectangle.getArea());
        }
        
        // Access interface constants
        System.out.println("\nGraphics constants:");
        System.out.println("PI: " + GraphicsConstants.PI);
        System.out.println("MAX_WIDTH: " + GraphicsConstants.MAX_WIDTH);
    }
    
    // Interface as method parameter type
    public static void processDrawable(Drawable drawable) {
        System.out.println("Processing drawable object:");
        drawable.draw();
        System.out.println("Area calculation: " + drawable.getArea());
    }
    
    // Return interface type
    public static Drawable createRandomShape() {
        if (Math.random() > 0.5) {
            return new Circle(Math.random() * 10);
        } else {
            return new Rectangle(Math.random() * 10, Math.random() * 10);
        }
    }
    
    public static void demonstrateInterfaceTypes() {
        System.out.println("\n=== INTERFACES AS TYPES ===\n");
        
        // Interface as parameter
        Circle circle = new Circle(3.0);
        processDrawable(circle);
        
        System.out.println("\nRandom shape creation:");
        Drawable randomShape = createRandomShape();
        processDrawable(randomShape);
    }
    
    public static void main(String[] args) {
        demonstrateInterfaces();
        demonstrateInterfaceTypes();
        
        System.out.println("\nðŸŽ¯ INTERFACES SUMMARY:");
        System.out.println("â€¢ Define contracts without implementations");
        System.out.println("â€¢ Support multiple inheritance (unlike classes)");
        System.out.println("â€¢ Enable polymorphism through contract guarantees");
        System.out.println("â€¢ Constants automatically public static final");
        System.out.println("â€¢ Perfect for loose coupling and dependency injection");
    }
}
```

## Modern Java Interfaces (Java 8+)

**Default methods and static methods in interfaces**

```java
// Java 8+ interface features: default and static methods
public class ModernInterfaces {
    
    interface Vehicle {
        void start();  // Abstract method
        void stop();   // Abstract method
        
        // Default method - provides implementation
        default void honk() {
            System.out.println("HONK! HONK!");
        }
        
        // Default method with logic
        default void performMaintenance() {
            System.out.println("Performing routine maintenance...");
            checkEngine();
            checkBrakes();
            System.out.println("Maintenance complete!");
        }
        
        // Private default method (Java 9+)
        default void checkEngine() {
            System.out.println("  Checking engine...");
        }
        
        default void checkBrakes() {
            System.out.println("  Checking brakes...");
        }
        
        // Static method
        static boolean needsLicence() {
            return true; // Most vehicles need licences
        }
        
        static void displayVehicleFacts() {
            System.out.println("Vehicle Facts:");
            System.out.println("â€¢ Vehicles provide transportation");
            System.out.println("â€¢ Most require licences to operate");
            System.out.println("â€¢ Regular maintenance is important");
        }
    }
    
    static class Car implements Vehicle {
        private String make;
        
        public Car(String make) {
            this.make = make;
        }
        
        @Override
        public void start() {
            System.out.println(make + " car starting... Vroom!");
        }
        
        @Override
        public void stop() {
            System.out.println(make + " car stopping... Screech!");
        }
        
        // Can override default method if desired
        @Override
        public void honk() {
            System.out.println(make + " car: BEEP! BEEP!");
        }
    }
    
    static class Motorcycle implements Vehicle {
        private String brand;
        
        public Motorcycle(String brand) {
            this.brand = brand;
        }
        
        @Override
        public void start() {
            System.out.println(brand + " motorcycle starting... Kickstart!");
        }
        
        @Override
        public void stop() {
            System.out.println(brand + " motorcycle stopping... Idle...");
        }
    }
    
    // Functional interface (Java 8+ feature)
    @FunctionalInterface
    interface Processor<T> {
        T process(T input); // Single abstract method
        
        // Default method
        default void displayResult(T result) {
            System.out.println("Processing result: " + result);
        }
        
        // Static method
        static <U> Processor<U> identity() {
            return input -> input; // Returns input unchanged
        }
    }
    
    public static void demonstrateModernInterfaces() {
        System.out.println("=== MODERN JAVA INTERFACES (Java 8+) ===\n");
        
        Car car = new Car("Toyota");
        Motorcycle bike = new Motorcycle("Honda");
        
        Vehicle[] vehicles = {car, bike};
        
        for (Vehicle vehicle : vehicles) {
            System.out.println("Vehicle demonstration:");
            vehicle.start();
            vehicle.honk();        // Default method - available to all
            vehicle.performMaintenance(); // Default method with logic
            vehicle.stop();
            System.out.println();
        }
        
        // Static interface methods
        System.out.println("Static interface information:");
        System.out.println("Vehicles need licence: " + Vehicle.needsLicence());
        Vehicle.displayVehicleFacts(); // Call static method
        
        // Functional interface demonstration
        System.out.println("\n=== FUNCTIONAL INTERFACE DEMO ===\n");
        
        // Lambda expression implementing functional interface
        Processor<String> stringProcessor = text -> "[PROCESSED] " + text.toUpperCase();
        
        String result = stringProcessor.process("hello world");
        stringProcessor.displayResult(result);
        
        // Using static method from functional interface
        Processor<String> identityProcessor = Processor.identity();
        String unchanged = identityProcessor.process("no change");
        identityProcessor.displayResult(unchanged);
    }
    
    public static void benefitsOfModernInterfaces() {
        System.out.println("\n=== BENEFITS OF MODERN INTERFACES ===\n");
        
        System.out.println("âœ… Default Methods:");
        System.out.println("   â€¢ Add new methods without breaking existing implementations");
        System.out.println("   â€¢ Provide common functionality across all implementers");
        System.out.println("   â€¢ Enable backward compatibility in APIs");
        
        System.out.println("\nâœ… Static Methods:");
        System.out.println("   â€¢ Utility methods related to the interface");
        System.out.println("   â€¢ Factory methods for creating implementations");
        System.out.println("   â€¢ Constant related helper methods");
        
        System.out.println("\nâœ… Functional Interfaces:");
        System.out.println("   â€¢ Single abstract method enables lambda expressions")
        System.out.println("   â€¢ Foundation for functional programming in Java");
        System.out.println("   â€¢ Essential for modern Java APIs (Stream, Optional, etc.)");
        
        System.out.println("\nðŸŽ¯ MODERN JAVA POWER:");
        System.out.println("Interfaces are no longer just contracts - they're active participants");
        System.out.println("in your class design, providing both structure AND behavior!");
    }
    
    public static void main(String[] args) {
        demonstrateModernInterfaces();
        benefitsOfModernInterfaces();
        
        System.out.println("\nðŸš€ ABSTRACT CLASSES & INTERFACES MASTERED:");
        System.out.println("Contract-based design patterns unlocked!");
        System.out.println("Enterprise Java architecture foundation established!");
    }
}
```

## Abstract Classes vs Interfaces

**Choosing the right contract-based design approach**

```java
// When to use abstract classes vs interfaces
public class AbstractVsInterface {
    
    // BASE CLASS EXAMPLE (nice to have, but can't force)
    static class BasicDevice {
        public void basicOperation() {
            System.out.println("Basic device operation");
        }
    }
    
    // Scenario 1: Abstract class - when you want to share behavior AND state
    static abstract class DatabaseConnection {
        protected String connectionString;
        protected boolean isConnected = false;
        
        public DatabaseConnection(String connectionString) {
            this.connectionString = connectionString;
        }
        
        // Abstract method - database-specific
        public abstract void executeQuery(String query);
        
        // Shared concrete behavior
        public void connect() {
            if (!isConnected) {
                System.out.println("Connecting to database: " + connectionString);
                isConnected = true;
                onConnect();
            }
        }
        
        public void disconnect() {
            if (isConnected) {
                System.out.println("Disconnecting from database");
                onDisconnect();
                isConnected = false;
            }
        }
        
        // Hook methods for subclasses
        protected void onConnect() {
            System.out.println("Connection established");
        }
        
        protected void onDisconnect() {
            System.out.println("Connection closed");
        }
    }
    
    static class MySQLConnection extends DatabaseConnection {
        public MySQLConnection(String url) {
            super("mysql://" + url);
        }
        
        @Override
        public void executeQuery(String query) {
            System.out.println("MySQL executing: " + query);
        }
    }
    
    // Scenario 2: Interfaces - when you only want to define contracts
    interface PaymentProcessor {
        boolean processPayment(double amount, String currency);
        String getProcessorName();
        double getTransactionFee(double amount);
        
        // Default implementation
        default boolean validateAmount(double amount) {
            return amount > 0 && amount <= 10000; // Reasonable limits
        }
    }
    
    interface Refundable {
        boolean processRefund(String transactionId, double amount);
        int getRefundWindowDays();
    }
    
    // Implementation combining multiple interfaces
    static class PayPalProcessor implements PaymentProcessor, Refundable {
        private String merchantId;
        
        public PayPalProcessor(String merchantId) {
            this.merchantId = merchantId;
        }
        
        @Override
        public boolean processPayment(double amount, String currency) {
            System.out.println("PayPal processing $" + amount + " " + currency);
            return true;
        }
        
        @Override
        public String getProcessorName() {
            return "PayPal";
        }
        
        @Override
        public double getTransactionFee(double amount) {
            return amount * 0.029 + 0.30; // 2.9% + 30 cents
        }
        
        @Override
        public boolean processRefund(String transactionId, double amount) {
            System.out.println("PayPal processing refund of $" + amount);
            return true;
        }
        
        @Override
        public int getRefundWindowDays() {
            return 180; // 180 days for PayPal
        }
    }
    
    public static void demonstrateUseCases() {
        System.out.println("=== WHEN TO USE ABSTRACT CLASSES VS INTERFACES ===\n");
        
        System.out.println("ðŸŽ¯ ABSTRACT CLASSES - Use when:");
        System.out.println("â€¢ Subclasses share common state (fields)");
        System.out.println("â€¢ Subclasses share common behavior implementations");
        System.out.println("â€¢ You need to define constructor logic");
        System.out.println("â€¢ IS-A relationship: Dog IS-A Animal (shared characteristics)");
        
        System.out.println("\nExample: Database connections share connection state and lifecycle:");
        DatabaseConnection db = new MySQLConnection("localhost:3306/mydb");
        db.connect();
        db.executeQuery("SELECT * FROM users");
        db.disconnect();
        
        System.out.println("\nðŸŽ¯ INTERFACES - Use when:");
        System.out.println("â€¢ Defining contracts without dictating implementation");
        System.out.println("â€¢ Multiple inheritance needed (Java's limitation with classes)");
        System.out.println("â€¢ Decoupling implementation from interface");
        System.out.println("â€¢ CAN-DO relationship: Logger CAN log, CAN serialize (different behaviors)");
        
        System.out.println("\nExample: Payment processors can do different things:");
        PaymentProcessor paypal = new PayPalProcessor("merchant123");
        double amount = 100.0;
        
        if (paypal.validateAmount(amount)) {
            paypal.processPayment(amount, "USD");
            System.out.println("Fee: $" + paypal.getTransactionFee(amount));
            
            // Check refund capability
            if (paypal instanceof Refundable) {
                Refundable refundProcessor = (Refundable) paypal;
                System.out.println("Refund window: " + refundProcessor.getRefundWindowDays() + " days");
            }
        }
    }
    
    public static void codeExamples() {
        System.out.println("\n=== QUICK REFERENCE GUIDE ===\n");
        
        System.out.println("ABSTRACT CLASS:");
        System.out.println("abstract class Entity {");
        System.out.println("    protected String id;");
        System.out.println("    public Entity(String id) { this.id = id; }");
        System.out.println("    public abstract void save();  // No implementation");
        System.out.println("    public void validate() { /* shared validation */ }");
        System.out.println("}");
        
        System.out.println("\nINTERFACE:");
        System.out.println("interface Validatable {");
        System.out.println("    boolean isValid();");
        System.out.println("    default void validate() { /* default impl */ }");
        System.out.println("}");
        
        System.out.println("\nIMPLEMENTATION DECISION:");
        System.out.println("â€¢ Abstract Class: Shared state + shared behavior = IS-A");
        System.out.println("â€¢ Interface: Contract only = CAN-DO relationship");
        
        System.out.println("\nMODERN JAVA TIP:");
        System.out.println("Use interfaces by default, abstract classes when state/behavior sharing is critical!");
    }
    
    public static void main(String[] args) {
        demonstrateUseCases();
        codeExamples();
        
        System.out.println("\nðŸ† CONTRACT-BASED DESIGN MASTERED:");
        System.out.println("Abstract Classes: Partial implementation blueprints");
        System.out.println("Interfaces: Pure contract guarantees"); 
        System.out.println("Together: Flexible, maintainable enterprise architectures!");
        System.out.println("\nNext: Generics - Type-safe, parameterized code!");
    }
}
```
