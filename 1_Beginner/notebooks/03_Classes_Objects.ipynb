## Object-Oriented Programming Basics

**OOP treats data and behavior as interconnected objects**

```java
// First understanding of classes and objects
public class FirstClass {
    
    public static void main(String[] args) {
        System.out.println("OOP organizes code around OBJECTS!");
        System.out.println("Each object combines DATA (fields) + BEHAVIOR (methods)");
        System.out.println();
        
        System.out.println("CLASS vs OBJECT examples:");
        System.out.println("Class 'Car' is a blueprint/recipe");
        System.out.println("Object 'myCar' is a specific car instance");
        System.out.println("Object 'yourCar' is a different car instance");
    }
}
```

## Class Fields (Instance Variables)

**Store data unique to each object instance**

```java
// Student class with instance fields
public class Student {
    
    // Instance variables (fields) - each student object has their own
    String name;
    int age;
    double gpa;
    String major;
    boolean isEnrolled;
    
    public static void main(String[] args) {
        // Create student objects
        Student alice = new Student();
        Student bob = new Student();
        
        // Set field values for each object
        alice.name = "Alice Johnson";
        alice.age = 20;
        alice.gpa = 3.7;
        alice.major = "Computer Science";
        alice.isEnrolled = true;
        
        bob.name = "Bob Smith";
        bob.age = 19;
        bob.gpa = 3.2;
        bob.major = "Biology";
        bob.isEnrolled = true;
        
        // Display student information
        System.out.println("Student 1:");
        System.out.println("  Name: " + alice.name);
        System.out.println("  Age: " + alice.age);
        System.out.println("  Major: " + alice.major);
        System.out.println("  GPA: " + alice.gpa);
        System.out.println("  Enrolled: " + alice.isEnrolled);
        
        System.out.println("\nStudent 2:");
        System.out.println("  Name: " + bob.name);
        System.out.println("  Age: " + bob.age);
        System.out.println("  Major: " + bob.major);
        System.out.println("  GPA: " + bob.gpa);
        System.out.println("  Enrolled: " + bob.isEnrolled);
        
        System.out.println("\nEach object maintains separate field values!");
    }
}
```
